```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data=pd.read_csv('C://Users//USER//Documents//teleconnect.csv')
```

```python
data.drop('customerID',axis=1,inplace=True)
```

```python
# Preprocessing
# Encode categorical variables
label_encoders = {}
categorical_columns = ['gender','Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 
                       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 
                       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 
                       'PaymentMethod', 'Churn']
```

```python
for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    data[column] = label_encoders[column].fit_transform(data[column])
```

```python
data.head()
```

```python
# Handle missing values in TotalCharges
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['TotalCharges'].fillna(data['TotalCharges'].mean(), inplace=True)
```

```python
# Features and target
X = data.drop('Churn', axis=1)
y = data['Churn']
```

```python
# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Define the existing model
model = Sequential()
model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the existing model
model.fit(X_train, y_train, epochs=50, batch_size=10, validation_split=0.2)
```

```python
# Evaluate the existing model
baseline_loss, baseline_accuracy = model.evaluate(X_test, y_test)
print(f"Baseline Model - Loss: {baseline_loss}, Accuracy: {baseline_accuracy}")
```

# Optimization Algorithms

```python
# Function to build and compile the model with a given optimizer
def build_compile_model(optimizer):
    model = Sequential()
    model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

optimizers = {
    'Adam': 'adam',
    'RMSprop': 'rmsprop',
    'SGD with Momentum': tf.keras.optimizers.SGD(momentum=0.9)
}
```

```python
# Dictionary to store results
results = {}

# Train and evaluate the model with different optimizers
for opt_name, opt in optimizers.items():
    print(f"\nTraining with {opt_name} optimizer")
    model = build_compile_model(opt)
    model.fit(X_train, y_train, epochs=50, batch_size=10, validation_split=0.2, verbose=0)
    loss, accuracy = model.evaluate(X_test, y_test)
    results[opt_name] = (loss, accuracy)
    print(f"{opt_name} - Loss: {loss}, Accuracy: {accuracy}")
```
